```python
import tensorflow as tf
from tensorflow import keras
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# Classifying handwritten digits using a Deep Learning Model.
```

The first challenge that we have is how do we feed the image as input to the neurons

So, to do that, we first convert the image to black and white so the pixel value may range from 0 (Black) to 255 (White). Then we represent the image as a 2D Matrix

Capture.PNG

Now, we convert 2D array to 1D array by arranging them into a single array matlab row1 then row 2 then row 3 and so on in a single array, and input each pixel to an input neuron in the neural network.

The problem we are trying to solve, will have images of 28x28 resolution, so we will have a 1D matrix having 784 pixel values.

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train[0].shape # Gives the rows and columns in the matrix.
```

```python
X_train[0] # This is the image data arranged in a 28 by 28 2D Array. First we need to convert it into a single matrix.
```

```python
plt.matshow(X_train[0]) # Displays an 2D array as a matrix. Colors of the image depends on magnitude at that location in matrix.
y_train[0]  # Our output is 5 and that matches the image. So, we have our dataset loaded successfully now.
```

```python
X_train.shape # Gives us the number of elements in the array, and the dimensions of the array.
```

```python
# Converting our image data from 2D Array to 1D array using numpy's reshape method.
# Reshape method takes in 2 parameters, the number of elements in the array, the new dimensions of the array.
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```python
X_train_flattened[0]  # Looking at how our image data looks like now, in an 1D array.
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape = (784,), activation='softmax')  # Here we created a network with 1 input layer and 1 output layer. No hidden layer.
])

model.compile(
    optimizer='adam', loss='sparse_categorical_crossentropy',metrics=['accuracy'] # This configures the model for training. So before we start training, we have to set our learning rate, other parameters in this function.
)

model.fit(X_train_flattened, y_train, epochs=5) # Training our model 
```

Scaling our data improves our machine learning model's accuracy. So we divide our image data by 255 to get a value between 0 and 1 for each pixel.

```python
X_train_flattened = X_train_flattened/255
X_test_flattened = X_test_flattened/255
```

```python
model.fit(X_train_flattened, y_train, epochs=5) # Again training our model with scaled image data. And we see improved results already.
```

```python
model.evaluate(X_test_flattened, y_test)  # Testing our model with test data.
```

```python
# Let's play around with the model a bit now.
plt.matshow(X_test[0])  # Printing first image from test dataset.
y_predicted = model.predict(X_test_flattened) # Predicting values from test dataset.
print(y_predicted[0])  # Gives us the predictions for the first image from dataset. It is an array of 10 numbers which is the probability value for each digit.
print(np.argmax(y_predicted[0])) # Gives us the index of the maximum value in the array. So, the index with highest probability is returned.
```

```python
# Creating a confusion matrix. Basically plotting the real values and guess values.
y_predicted_labels = [np.argmax(i) for i in y_predicted] # Doing argmax for each array in predicted arrays.
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
print(cm)
```

```python
# Displaying Confusion Matrix visually in a better way using Seaborn
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted Value')
plt.ylabel('Real Value')
```

Now, let's create another model with a hidden layer unlike the previous model. So that it gives us better results.

```python
model2 = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),  # This is the hidden layer with 100 neurons.
    keras.layers.Dense(10, activation='softmax')
])

model2.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model2.fit(X_train_flattened, y_train, epochs=5)
model2.evaluate(X_test_flattened, y_test)
```

```python
# Now, everytime I want to input image data into the model, I have to first flatten the data (i.e Convert it to 1D array).
# Instead of that, we can use keras.Flatten as the input layer to do the same.
model3 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),  # This is the input layer. It flattens the 28x28 array to single array and takes in inputs.
    keras.layers.Dense(64, activation='relu'),  # First hidden layer with 64 neurons
    keras.layers.Dense(32, activation='relu'),  # Another hidden layer with 16 neurons
    keras.layers.Dense(16, activation='relu'),  # Another hidden layer with 32 neurons
    keras.layers.Dense(10, activation='softmax')  # Output layer with 10 neurons
])

model3.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model3.fit(X_train, y_train, epochs=5)  # Notice how I'm not passing the flattened arrays here.
model3.evaluate(X_test, y_test) # Here also, we're not passing the flattened arrays.
```
